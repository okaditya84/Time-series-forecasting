```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('extracted important\AshokNagar_extracted.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.interpolate(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['Temp', 'RF'], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Iterate over the columns of the dataframe
for column in df.columns:
    # Plot the column against the index
    plt.plot(df.index, df[column], label=column)

# Set the x-axis label
plt.xlabel('Year')

# Set the y-axis label
plt.ylabel('Parameter')

# Set the title of the plot
plt.title('Line Graph of Parameters Yearly')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
